```python
import pandas as pd
import matplotlib.pyplot as plt
import pylab
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
def colors(flag):
  color = {-1: "#000000",
            0: "#FF0000",
            1: "#FF7400",
            2: "#009999",
            3: "#00CC00",
            4: "#4811AE",
            5: "#2C0571",
            6: "#FFD373",
            7: "#A64B00",
            8: "#5CCCCC"}
  col = []
  for i in flag:
    if i <= 8:
      col.append(color[i])
    else:
      col.append("FBF234")
  return col
```

```python
df = pd.read_csv("dataset.csv")
df.head()
df.shape
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 0:2]
y = df.iloc[:, 2]
```

```python
kmean = []
clu = [3, 6, 9]
eps = [0.4, 0.65, 1]

for i in clu:
  kmeans = DBSCAN(min_samples=i, eps=0.5).fit(X)
  kmean.append(kmeans.labels_)

for i in eps:
  kmeans = DBSCAN(min_samples=6, eps=i).fit(X)
  kmean.append(kmeans.labels_)
```

```python
fig, axs = plt.subplots(nrows= 6 , ncols= 2 , figsize=(15, 30))
axs[0][0].set_title('Исходные данные')
axs[0][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[0][1].set_title('Полученные данные при значении: 3')
axs[0][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[0]))
axs[1][0].set_title('Исходные данные')
axs[1][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[1][1].set_title('Полученные данные при значении: 6')
axs[1][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[1]))
axs[2][0].set_title('Исходные данные')
axs[2][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[2][1].set_title('Полученные данные при значении: 9')
axs[2][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[2]))
axs[3][0].set_title('Исходные данные')
axs[3][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[3][1].set_title('Полученные данные при значении: 0.3')
axs[3][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[3]))
axs[4][0].set_title('Исходные данные')
axs[4][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[4][1].set_title('Полученные данные при значении: 0.7')
axs[4][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[4]))
axs[5][0].set_title('Исходные данные')
axs[5][0].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(y))
axs[5][1].set_title('Полученные данные при значении: 1')
axs[5][1].scatter(X.iloc [:, 0],X.iloc [:, 1], color = colors(kmean[5]))
```
