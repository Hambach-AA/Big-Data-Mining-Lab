```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = [random.normalvariate(1000, 50) for i in range(200)]
dataset_ = [random.normalvariate(700, 50) for i in range(5)]
dataset = dataset + dataset_

dataset2 = [random.normalvariate(1000, 50) for i in range(200)]
dataset_ = [random.normalvariate(700, 50) for i in range(5)]
dataset2 = dataset2 + dataset_
```

```python
def gistogram(dataset):
  dataset.sort()
  test = []
  for i in range(50,2050,50):
    test.append(0)
    for j in range(len(dataset)):
      if dataset[j] < i and dataset[j] > i - 50:
        test[-1] += 1
  test2 = test.copy()
  for i in range(len(test)):
    if test2[i] != 0:
      test2[i] = 1/(100*test2[i]/len(dataset))
  color = []
  for i in range(len(test2)):
      if test2[i] > 0 and test2[i] < 1:
        for j in range(test[i]):
          color.append("b")
      elif test2[i] > 0 and test2[i] > 1:
        for j in range(test[i]):
          color.append("r")
  return color
```

```python
def maximum_likelihood(dataset):
  color = []
  theta_A = dataset / np.sum(dataset)
  for i in theta_A:
    if i < 0.004:
      color.append("r")
    else:
      color.append("b")
  return color
```

```python
plt.scatter(x = dataset, y = [i*0 for i in range(len(dataset))], color=gistogram(dataset))
```

```python
plt.scatter(x = dataset, y = [i*0 for i in range(len(dataset))], color=maximum_likelihood(dataset))
```

```python
plt.scatter(x = dataset, y = [i*0 for i in range(len(dataset))], color=gistogram(dataset2))
```

```python
plt.scatter(x = dataset, y = [i*0 for i in range(len(dataset))], color=maximum_likelihood(dataset2))
```
